#2. One Hot Encoding

```python
d = {'ID':[11101, 11102, 11103, 11104, 11105, 11106],
     'Nama Produk':['Tomat', 'Apel', 'Pisang', 'Semangka', 'Melon', 'Jeruk']}
df = pd.DataFrame(data=d)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
y = OneHotEncoder(sparse=False)
y = y.fit_transform(df[['Nama Produk']])
y
```

```python
# Mengubah bentuk array ke dataframe
data_y = pd.DataFrame(y)
data_y
```

```python
# Memberi nama header column dengan nama produk
header = df['Nama Produk'].sort_values()
data_y.columns = 'Produk_' + header
data_y
```

```python
data_join = df.join(data_y)
data_join
```
